```python
# Importando Librerías
import re
import json
import pickle
import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

from classes.game import Game
import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_columns', None)
```

```python
df_2021 = pd.read_pickle('data/2021.pkl')
df_2021
```

```python
matches_links = df_2021.Link.values
```

```python
anio_2021 = Game()
```

```python
stats_column = []
for link in matches_links:

    anio_2021.get_web(link)

    stats_column.append(anio_2021.game_stats())
```

```python
stats_column
```

```python
df_2021['stats'] = pd.Series(stats_column)
```

```python
df_2021
```

```python
def clean_stats_column(dictionary):
    local_visit = {}
    for key, value in dictionary.items():
        local_visit[key+' LOCAL'] = value[0]
        local_visit[key+' VISITANTE'] = value[1]

    return local_visit
```

```python
df_2021['stats'] = df_2021['stats'].apply(lambda x: clean_stats_column(x))
```

```python
df_2021
```

```python
df_2021_clean = pd.concat([df_2021.drop(['stats'], axis=1), df_2021['stats'].apply(pd.Series)], axis=1)
df_2021_clean
```
